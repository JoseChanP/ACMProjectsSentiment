```python
import pandas as pd
import nltk
import string
nltk.download('vader_lexicon')
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score

headlines_df = pd.read_csv('final.csv')
headlines_df.head()
```

```python
headlines_df['Title'] = headlines_df['Title'].str.lower().str.translate(str.maketrans('','', string.punctuation)) # Convert 'text' column to string data type
headlines_df['Title'] = headlines_df['Title'].apply(word_tokenize)
stopwords = set(stopwords.words('english'))
headlines_df['Title'] = headlines_df['Title'].apply(lambda x: [word for word in x if word not in stopwords])
lemmatizer = WordNetLemmatizer()
headlines_df['Title'] = headlines_df['Title'].apply(lambda x: [lemmatizer.lemmatize(token) for token in x])
headlines_df
```
